# 2) Exploratory Data Analysis (EDA)

```python
#Data Information
df.info()

# 3 columns have null statement, 5 columns are integer, 15 columns are string.
```

```python
# Data Cleaning and Transforming


df['Profit'] = df['Profit'].replace({'\$': ''}, regex=True)
df['Profit'] = df['Profit'].apply(lambda x: '-' + x[1:-1] if x.startswith('(') and x.endswith(')') else x)
df['Profit'] = df['Profit'].replace({',': ''}, regex=True).astype(float)
df['Profit'] = df['Profit'].astype(float)

df['Total Revenue'] = df['Total Revenue'].replace({'\$': ''}, regex=True)
df['Total Revenue'] = df['Total Revenue'].replace({',': ''}, regex=True).astype(float)
df['Total Revenue'] = df['Total Revenue'].astype(float)

df['Unit Cost'] = df['Unit Cost'].replace({'\$': ''}, regex=True)
df['Unit Cost'] = df['Unit Cost'].replace({',': ''}, regex=True).astype(float)
df['Unit Cost'] = df['Unit Cost'].astype(float)

df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Delivery_Date'] = pd.to_datetime(df['Delivery_Date'])
df['Customer_BirthDate'] = pd.to_datetime(df['Customer_BirthDate'])

# Dates transformed in the ISO 8601 format. The negative values which are shown in paranthesis transformed into minus display. Numerical object data transformed into float.
```

```python
df.head()
```

#### Eliminating the Duplicated Data by Aggregating Method

```python
# Duplicated variables are eliminated by aggregating data by CustomerID and OrderID.

df = df.groupby(['Customer ID', 'Order ID'], as_index=False).agg({
    'Total Revenue': 'sum',
    'Profit': 'sum',
    'Unit Cost': 'sum',
    'Quantity': 'sum',
    'Discount': 'first',
    'Days to Delivery': 'first',
    'Order_Date': 'first',
    'Delivery_Date': 'first',
    'OrderTypeLabel': 'first',
    'CustomerCountryLabel': 'first',
    'Customer_BirthDate': 'first',
    'Customer_Group': 'first',
    'Customer_Type': 'first',
    'Loyalty Num': 'first',
    'Customer ID': 'first',
    'City' : 'first',
    'Continent' : 'first',
    'Postal_Code' : 'first',
    'State_Province' : 'first',
    'Order ID' : 'first'

})
print(df.shape)
```

# 3) Feature Engineering

```python
#Aggregating Numerical data into the columns (CustomerID, Frequency, Recency, CLV, AverageUnitCost,CustomerAge)

# 1. Frequency: Calculate how often a customer makes a purchase

frequency = df.groupby('CustomerID')['OrderID'].count().reset_index()
frequency.columns = ['CustomerID', 'Frequency']

# 2. Recency: Calculate how recently a customer has made a purchase

latest_order_date = df.groupby('CustomerID')['Order_Date'].max().reset_index()
latest_order_date.columns = ['CustomerID', 'Most Recent Order Date']
today = df['Order_Date'].max()
latest_order_date['Recency'] = (today - latest_order_date['Most Recent Order Date']).dt.days

# 3. Customer Lifetime Value (CLV): Total revenue for each customer

clv = df.groupby('CustomerID')['TotalRevenue'].sum().reset_index()
clv.columns = ['CustomerID', 'CLV']

# 4. Average Unit Cost: Calculate the average cost per unit ordered for each customer

df['TotalUnitCost'] = df['UnitCost'] * df['Quantity']
average_unit_cost = df.groupby('CustomerID')['TotalUnitCost'].sum() / df.groupby('CustomerID')['Quantity'].sum()
average_unit_cost = average_unit_cost.reset_index()
average_unit_cost.columns = ['CustomerID', 'AverageUnitCost']

# 5. Customer Age: Calculate the age of the customer based on their birthdate

today = datetime.today()
df['CustomerAge'] = today.year - df['Customer_BirthDate'].dt.year
df['CustomerAge'] -= ((today.month < df['Customer_BirthDate'].dt.month) |
                        ((today.month == df['Customer_BirthDate'].dt.month) & (today.day < df['Customer_BirthDate'].dt.day))).astype(int)

# Merge all the new features together into a final customer feature dataframe

customer_features = frequency.merge(latest_order_date[['CustomerID', 'Recency']], on='CustomerID')
customer_features = customer_features.merge(clv[['CustomerID', 'CLV']], on='CustomerID')
customer_features = customer_features.merge(average_unit_cost[['CustomerID', 'AverageUnitCost']], on='CustomerID').astype(int)
customer_features = customer_features.merge(df[['CustomerID', 'CustomerAge']].drop_duplicates(), on='CustomerID')

# Display the new customer features dataframe
print(customer_features.head(10))
```

# 4) Outlier Detection and Elimination

```python
#Used z-scores to detect Outliers

df_z_scores = customer_features.apply(zscore)
threshold = 3
outliers = (df_z_scores.abs() > threshold).any(axis=1)

# Filter out the outliers from the DataFrame

df_cleaned_zscore = customer_features[~outliers]

# Results after Extracting the Outliers

print(f'Outliers Numbers: {len(customer_features) - len(df_cleaned_zscore)}')
print(f'Number of rows before outlier removal: {len(customer_features)}')
print(f'Number of rows after outlier removal: {len(df_cleaned_zscore)}')
```

# 5) Feature Scaling

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), customer_features.columns),  # Scale numerical columns
        ('cat', OneHotEncoder(), customer_features.columns)  # One-hot encode categorical columns
    ])

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor)
])


X_processed = pipeline.fit_transform(customer_features)

```

<a id = "1"></a></br>

# 6) Optimum K Number Detection

```python
inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_processed)
    inertia.append(kmeans.inertia_)

# Inertia Graphics
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method: Inertia vs K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.show()

#Observed that there is an elbow at 5
```

```python
# K is chosen as 5
k = 5
kmeans = KMeans(n_clusters=k, init='random', random_state=42)

kmeans.fit(X_processed)

labels = kmeans.labels_

# Add the cluster labels to the DataFrame

customer_features['Cluster'] = labels

```
